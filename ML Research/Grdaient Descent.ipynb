### Problem 1

```python
def func(x, b):
    return (x - b) ** 2

def d_func(x, b):
    return 2 * (x - b)

def gradient_descent(lr, x, b):
    history = []  
    while abs(d_func(x, b)) > 0.0001:
        fx = func(x, b)
        history.append((x, fx))
        x = x - lr * d_func(x, b)
    history.append((x, func(x, b)))  
    return x, history

# Run gradient descent
x_initial = 0
b_value = 5
learning_rate = 0.1

min_x, steps = gradient_descent(learning_rate, x_initial, b_value)

print("Minimum value at x =", min_x)
print("\nDescent steps:")
for i, (x_val, fx_val) in enumerate(steps):
    print(f"Step {i + 1}: x = {x_val:.6f}, f(x) = {fx_val:.6f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Extract x and f(x) values for plotting
x_vals, f_vals = zip(*steps)

# Prepare the full curve of the function
x_curve = np.linspace(min(x_vals) - 1, b_value + 1, 300)
f_curve = func(x_curve, b_value)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_curve, f_curve, label='f(x) = (x - b)^2', color='gray')
plt.plot(x_vals, f_vals, marker='o', linestyle='-', color='blue', label='Gradient Descent Path')
plt.title('Gradient Descent on f(x) = (x - b)^2')
plt.xlabel('x value')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return x * np.log(x) if x > 0 else float('inf')

def d_func(x):
    return 1 + np.log(x) if x > 0 else float('inf')

def gradient_descent(lr, x):
    history = []
    while x > 0 and abs(d_func(x)) > 0.0001:
        fx = func(x)
        history.append((x, fx))
        x = x - lr * d_func(x)
        if x <= 0:  # Prevent leaving the domain
            break
    if x > 0:
        history.append((x, func(x)))
    return x, history

x_initial = 2
learning_rate = 0.1

min_x, steps = gradient_descent(learning_rate, x_initial)

print("Minimum value at x =", min_x)
for i, (x_val, fx_val) in enumerate(steps):
    print(f"Step {i + 1}: x = {x_val:.6f}, f(x) = {fx_val:.6f}")

# Plotting
x_vals, f_vals = zip(*steps)
x_curve = np.linspace(0.1, 4, 300)
f_curve = x_curve * np.log(x_curve)

plt.figure(figsize=(10, 6))
plt.plot(x_curve, f_curve, label='f(x) = x log x', color='gray')
plt.plot(x_vals, f_vals, marker='o', linestyle='-', color='blue', label='Gradient Descent Path')
plt.title('Gradient Descent on f(x) = x log x')
plt.xlabel('x value')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("xlogx_plot.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function and gradient
def func(x, A, b):
    diff = x - b
    return np.linalg.norm(A @ diff) ** 2

def grad(x, A, b):
    return 2 * A.T @ A @ (x - b)

# Gradient descent
def gradient_descent(lr, x_init, A, b, tol=1e-6):
    x = x_init
    history = []
    while np.linalg.norm(grad(x, A, b)) > tol:
        history.append((x.copy(), func(x, A, b)))
        x -= lr * grad(x, A, b)
    history.append((x.copy(), func(x, A, b)))
    return x, history

# Define A, b, and initial x
np.random.seed(0)
A = np.random.randn(3, 2)  # A: 3x2 matrix
b = np.array([1.0, 2.0])   # target vector
x0 = np.array([0.0, 0.0])  # initial guess
lr = 0.1

# Run
min_x, steps = gradient_descent(lr, x0, A, b)

# Output
print("Minimum found at x =", min_x)
for i, (x, fx) in enumerate(steps[:22]):  # Show first few steps
    print(f"Step {i + 1}: x = {x}, f(x) = {fx:.6f}")
# Plot contour and path
import numpy as np
X, Y = np.meshgrid(np.linspace(-2, 3, 100), np.linspace(-2, 3, 100))
Z = np.array([[func(np.array([x, y]), A, b) for x in X[0]] for y in Y[:, 0]])

plt.figure(figsize=(6, 4))
plt.contour(X, Y, Z, levels=30, cmap='viridis')
path = np.array([p[0] for p in steps])
plt.plot(path[:, 0], path[:, 1], 'ro-', label='GD Path')
plt.scatter(b[0], b[1], color='green', marker='x', label='Target b')
plt.title("Gradient Descent Path for $f(x) = \\|A(x - b)\\|^2$")
plt.xlabel('x[0]')
plt.ylabel('x[1]')
plt.legend()
plt.grid(True)
plt.show()
```
